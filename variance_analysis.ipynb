Variance Analysis

the formula for variance futures is:

$$f^{VA}t(T) = \frac{10,000}{T-t_0} \left( \int{t_0}^t V_u du + a^(T-t) + b^(T-t)V_t \right)$$

where:

$a^*(T-t) = \theta \left[ (T-t) - \frac{1-e^{-\lambda(T-t)}}{\lambda} \right]$
$b^*(T-t) = \frac{1-e^{-\lambda(T-t)}}{\lambda}$
Here's how each parameter affects the variance futures price:

1. Mean Reversion Speed (λ)
Controls how quickly squared volatility reverts to its long-term mean
Higher λ means faster mean reversion
For a given T-t, higher λ means:
When Vt > θ: higher futures price (faster decay to mean)
When Vt < θ: lower futures price (faster rise to mean)
The term structure flattens more quickly with higher λ
2. Long-term Mean Level (θ)
The level that squared volatility tends toward in the long run
Directly impacts the a* term
Higher θ leads to linearly higher variance futures prices
For longer maturities, the impact of θ dominates
3. Current Squared Volatility (Vt)
Starting point for the volatility process
Impact decreases with maturity due to mean reversion
The impact is scaled by b*(T-t), which decreases with maturity
4. Accrued Variance
Has a linear effect on the futures price
Its impact is scaled by 1/(T-t₀)
Higher accrued variance → higher variance futures price
5. Time to Maturity (T-t)
Affects both a* and b* terms
For Vt = θ: variance futures price is nearly constant (flat term structure)
For Vt > θ: downward sloping term structure
For Vt < θ: upward sloping term structure
6. Total Time Span (T-t₀)
Acts as a scaling factor for the entire formula
Longer T-t₀ means less impact from the same accrued variance
The differences between variance futures and VIX futures are notable:

Variance futures have a closed-form solution, while VIX futures require numerical integration
Variance futures directly price realized variance, while VIX futures price the square root of expected variance
The convexity effect (Jensen's inequality) makes VIX futures behave differently from variance futures
